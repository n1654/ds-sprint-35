# Проект - поиск по изображениям

## Описание проекта

Вы работаете в фотохостинге для профессиональных фотографов «Со Смыслом» (“With Sense”). 

Ваши пользователи размещают свои фотографии на хостинге и сопровождают их полным описанием: указывают место съёмок, модель камеры и т. д. 

Отличительная особенность сервиса — описание: его может предоставить не только тот, кто размещает фотографию, но и другие пользователи портала.

Например, для этой фотографии



описание выглядит так:

⎢ A hiker poses for a picture in front of stunning mountains and clouds. 

Ваш отдел занимается экспериментом по разработке поиска референсных фотографий для фотографов. Суть поиска заключается в следующем: пользователь сервиса вводит описание нужной сцены. Например, такое:

⎢ A man is crossing a mountain pass on a metal bridge.

Сервис выводит несколько фотографий с такой же или похожей сценой.



Чтобы эксперимент получил право на жизнь, нужно защитить его перед руководителем компании.
Для защиты необходимо презентовать так называемый PoC (Proof of Concept, Проверка концепции) — продемонстрировать, что такой проект практически осуществим. 
Вам поручено разработать демонстрационную версию поиска изображений по запросу.
Для демонстрационной версии нужно выбрать лучшую  модель, которая получит векторное представление изображения, векторное представление текста, а на выходе выдаст число от 0 до 1 — и покажет, насколько текст и картинка подходят друг другу. 
На основе лучшей модели можно будет собрать предварительную версию продукта, которую вы покажете руководителю компании.

Юридические ограничения

В некоторых странах, где работает компания With Sense, действуют ограничения по обработке изображений: поисковым сервисам и сервисам, предоставляющим возможность поиска, запрещено без разрешения родителей или законных представителей предоставлять любую информацию, в том числе, но не исключительно, текстов, изображений, видео и аудио, содержащие описание, изображение или запись голоса детей. Ребёнком считается любой человек, не достигший 16-ти лет.
В вашем сервисе строго следуют законам стран, в которых работают. Поэтому при попытке посмотреть изображения, запрещённые законодательством, вместо картинок показывается дисклеймер:

⎢ This image is unavailable in your country in compliance with local laws.

### Суммарная оценка

Создадим итоговую экспертную оценку.

Сформирую два класса:
 - 1 - изображение соответствует описанию. Критерий: два и более эксперта поставили оценку 3 или 4.
 - 0 - изображение не соответствует описанию. Критерий: все остальные

```python
def estimate_exp(exp_1, exp_2, exp_3):
    # normalized average
    count = (exp_1 + exp_2 + exp_3) / 3 / 4
    return count

print(estimate_exp(4,4,4))
print(estimate_exp(4,3,4))
print(estimate_exp(4,1,4))
print(estimate_exp(4,0,0))
print(estimate_exp(2,3,4))
print(estimate_exp(0,1,1))
print(estimate_exp(0,0,0))
```

Я вижу, что для формирования целевого признака была выбрана стратегия бинаризации экспертных оценок. С технической точки зрения функция `estimate_exp` реализована верно, однако концептуально такой подход приводит к потере значительной части информации. 

Исходная шкала от 1 до 4 содержит в себе важные «полутона»: она позволяет отличить полное отсутствие соответствия от частичного, а частичное — от идеального. При превращении этих оценок в жесткие 0 и 1 мы лишаем модель возможности ранжировать результаты по степени релевантности, что является критически важным для поисковой системы. 

Я считаю, что в рамках данного проекта более оправдано использовать задачу регрессии. Если агрегировать оценки экспертов (например, через среднее арифметическое) и перевести их в диапазон $[0, 1]$, модель сможет улавливать более тонкие семантические связи между текстом и изображением.

Но если ты хочешь все-таки использовать классификацию (например, из-за метрики ROC-AUC, которая показывает, насколько правильно ранжированы подходящие и неподходящие пары), то стоит поискать способ использовать все крауд оценки.
```python
def filter_stop_word(text):
    if pd.isna(text):
        return False
    
    patterns = PATTERNS
    
    for pattern in patterns:
        if re.search(pattern, text, re.IGNORECASE):
            return True
    return False


text = "Four men playing basketball , two from each team ."
print(f"{text}: {filter_stop_word(text)}")

text = "Little kids stand together and raise their hands in the air ."
print(f"{text}: {filter_stop_word(text)}")

text = "A little boy is running towards a group of people ."
print(f"{text}: {filter_stop_word(text)}")

text = "two guy standing on horses , while a crowd watches"
print(f"{text}: {filter_stop_word(text)}")
```

## Векторизация изображений

```python
# Проверяем наличие GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Используется устройство: {device}")
```

```python
# Загружаем предобученную модель ResNet-18
img_model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
```

```python
layers = list(img_model.children())
```

```python
# Удаляем последний полносвязный слой (Linear)
img_model = nn.Sequential(*layers)[:-1]
img_model = img_model.to(device)

# режим оценки
img_model.eval();
```

```python
# Стандартные трансформации для ImageNet
img_transform = transforms.Compose([
    transforms.Resize(256),           # Изменяем размер до 256
    transforms.CenterCrop(224),       # Центральный кроп до 224x224 (требование ResNet)
    transforms.ToTensor(),            # Преобразуем в тензор [0, 1]
    transforms.Normalize(             # Нормализуем с параметрами ImageNet (https://docs.pytorch.org/vision/stable/models/generated/torchvision.models.quantization.resnet18.html#torchvision.models.quantization.ResNet18_QuantizedWeights)
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
def extract_image_embedding(image_path, model, transform, device):
    try:
        image = Image.open(image_path).convert('RGB')

        image_tensor = transform(image).unsqueeze(0)
        image_tensor = image_tensor.to(device)
        
        with torch.no_grad():
            embedding = img_model(image_tensor)
        
        embedding = embedding.squeeze().cpu().numpy()
        
        return embedding
    
    except Exception as e:
        print(f"Ошибка при обработке {image_path}: {e}")
        return None
```

```python
def vectorize_dataset(df, path, image_column, model, transform, device):
    embeddings = []
    
    for idx, row in tqdm(df.iterrows(), total=len(df), desc="Векторизация изображений"):
        image_path = path + row[image_column]
        embedding  = extract_image_embedding(image_path, img_model, transform, device)
        
        if embedding is not None:
            embeddings.append(embedding)
        else:
            embeddings.append(np.zeros(512))
    
    return np.array(embeddings)
```

```python
# Применяем функцию к датафрейму
image_embeddings = vectorize_dataset(
    train_data_merged_filtered,
    TRAIN_IMAGES,
    'image_origin', 
    img_model, 
    img_transform, 
    device
)
```

```python
# добавляем эмбеддинги в датафрейм
train_data_merged_filtered['image_embedding'] = list(image_embeddings)

print(f"Размер датафрейма: {train_data_merged_filtered.shape}")
```

Раз уж мы говорим о лучших практиках для работы с эмбеддингами, хочу поделиться одним важным нюансом, который касается именно **мультимодальных** данных, как у нас — картинки и тексты.

Твой текущий подход — через `StandardScaler` — абсолютно валиден для классических табличных данных. Но для эмбеддингов он может быть неоптимален. `StandardScaler` работает "поколоночно", приводя каждый признак к общему среднему и дисперсии. 


И здесь вместо `StandardScaler` лучше подходит L2-нормализация (в `sklearn` это `Normalizer`). Этот метод масштабирует каждый вектор (каждую строку целиком) к единичной длине. Он убирает влияние "магнитуды" вектора (которая часто является шумом) и оставляет только его **направление** в многомерном пространстве — а именно оно и несет основную семантическую информацию.
    
Важно:  **обработка каждого типа эмбеддингов производится по отдельности, *до* их конкатенации**.
    
Таким образом, более продвинутый пайплайн мог бы выглядеть так:
1.  Получаем матрицу эмбеддингов для изображений `X_img`.
2.  Получаем матрицу эмбеддингов для текстов `X_text`.
3.  Применяем `Normalizer()` к `X_img` и `X_text` **по отдельности**.
4.  Конкатенируем уже нормализованные матрицы.

Такой подход гарантирует, что и визуальные, и текстовые признаки вносят в модель сопоставимый вклад, не искажая при этом структуру каждого из исходных пространств. Это знание будет очень полезно тебе в будущих проектах.

## Модели нейронной сети

```python
def create_nn_model(input_dim, seed=RANDOM_STATE):
    """
    Создает нейронную сеть с фиксированными случайными инициализациями
    """

    model = Sequential([
        Dense(
            256, 
            activation='relu', 
            input_shape=(input_dim,),
            kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
        ),
        Dropout(0.4),
        
        Dense(
            128, 
            activation='relu',
            kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed+1)
        ),
        Dropout(0.3),
        
        Dense(
            64, 
            activation='relu',
            kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed+2)
        ),
        Dropout(0.2),
        
        Dense(
            1, 
            activation='sigmoid',
            kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed+3)
        )
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss='mse',
        metrics=['mse', 'mae']
    )
    
    return model
```

```python
# Создание модели
input_dim = X_train.shape[1]
nn_model  = create_nn_model(input_dim)

# Callbacks для лучшего обучения
early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=15, 
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=7,
    min_lr=0.00001,
    verbose=1
)
```

### Функция поиска

```python
def search_pair_nn(nn_model, txt_model, query, df_images,
                   normalizer_image, normalizer_text):

    if filter_stop_word(query):
        return "content cannot be determined or prohibited", 0, 0, 0
    
    text_embedding = txt_model.encode(query)
    
    image_embs = np.vstack(df_images['image_embedding'].values)
    
    image_embs_scaled = normalizer_image.transform(image_embs)
    text_embs_scaled  = normalizer_text.transform(text_embedding.reshape(1, -1))
    
    text_embs_normalized = np.tile(text_embs_scaled, (image_embs_scaled.shape[0], 1))
    
    combined_normalized = np.hstack([image_embs_scaled, text_embs_normalized])
    
    y_pred = nn_model.predict(combined_normalized, verbose=0).flatten()

    all_predictions = y_pred
    
    best_index = np.argmax(all_predictions)
    best_prediction = all_predictions[best_index]
    
    df_images['nn_prediction'] = all_predictions
    
    return "content - ok", best_index, best_prediction, all_predictions
```
